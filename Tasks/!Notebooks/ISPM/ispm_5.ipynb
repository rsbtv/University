4. ПОЯСНИТЕЛЬНАЯ ЗАПИСКА

4.1. Обоснование структуры календарного плана

Структура календарного плана отражает полный жизненный цикл исследовательско‑прикладного ML‑проекта: от анализа требований и предметной области до разработки, оптимизации и оформления результатов в виде выпускной квалификационной работы.
Выделение отдельного этапа «Проектирование архитектуры и методики обучения» обусловлено спецификой архитектуры HyenaPixel, использующей оператор Hyena для моделирования длинных свёрток с глобальным рецептивным полем и неявной параметризацией фильтров, что требует более тщательного архитектурного проектирования по сравнению с классическими CNN‑подходами.

Этап «Реализация и обучение моделей» разложен на несколько взаимосвязанных подпроцессов: инфраструктура данных (препроцессинг ДЗЗ), реализация модели HyenaPixel, реализация самоконтролируемых задач и проведение fine‑tuning для сегментации. Такое разбиение позволяет параллелить часть работ и при этом сохранять логическую зависимость критических задач (обучение и подготовка стабильного чекпоинта) от результатов проектирования.  

Фаза «Оптимизация инференса и экспериментальные исследования» выделена отдельно, так как в тематике работы ключевым является не только качество сегментации, но и эффективность инференса: достижение субквадратичной сложности и уменьшение времени обработки больших сцен ДЗЗ при сохранении качества.
Заключительный этап «Тестирование, документирование и подготовка защиты» обеспечивает перевод полученных научно‑практических результатов в форму, необходимую для защиты дипломной работы, включая формализацию экспериментов и интерпретацию результатов.  

4.2. Методология анализа рисков

Методология анализа рисков базируется на общепринятых подходах управления проектными рисками: идентификация потенциальных угроз, качественная оценка вероятности и влияния, формирование ранжированного перечня рисков и определение стратегий реагирования.
Особое внимание уделено рискам, типичным для проектов машинного обучения: качеству и объёму данных, вычислительным ограничениям, сложности архитектуры моделей и воспроизводимости экспериментов, а также организационным рискам, связанным со сроками выполнения дипломного проекта.  

Для каждого риска выбрана стратегия реагирования (преимущественно «снижение») и сформулированы конкретные превентивные меры, что обеспечивает практическую применимость разработанной системы управления рисками и её непосредственную связь с планом работ (например, резервы по времени, использование аугментаций и референс‑кода, фиксация конфигураций экспериментов).

4.3. Принципы определения контрольных точек

Контрольные точки определены таким образом, чтобы каждая из них фиксировала завершение содержательно значимого этапа: утверждение концепции и требований, завершение проектирования архитектуры, получение базовой обученной модели, завершение оптимизаций и экспериментов, готовность текста и презентационных материалов.
Критерии достижения сформулированы по принципу SMART: они конкретны, измеримы (через метрики качества, наличие документов, готовность артефактов), достижимы в рамках установленного календарного плана, релевантны целям проекта и ограничены по времени зафиксированными датами.  

Связь контрольных точек с календарным планом двусторонняя: с одной стороны, завершение групп задач по диаграмме Ганта приводит к достижению соответствующей вехи; с другой стороны, приближение дат вех служит ориентиром для мониторинга выполнения работ и актуализации плана при возникновении рисков и отклонений от первоначального графика.
