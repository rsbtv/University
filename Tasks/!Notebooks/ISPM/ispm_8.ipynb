# **Задание №8. Разработка пользовательских интерфейсов (прототипы интерфейса)**

## 1. ИНФОРМАЦИОННАЯ АРХИТЕКТУРА

### 1.1. Карта навигации (структура системы)

```
Роли:
    ML‑инженер
    Аналитик ДЗЗ
    Администратор

Глобальная структура (верхнее меню + левое боковое):
    Главная
        Обзор проекта (KPI, последние эксперименты, последние инференсы)
    Датасеты и сцены
        Списки датасетов
        Карточка датасета
        Список сцен
        Карточка сцены (изображение + метаданные + разметка)
    Инференс
        Запустить инференс
        Список запусков
        Карточка запуска (маска + метрики)
    Администрирование
        Пользователи и роли
        Настройки системы
```

### 1.2. Пользовательские пути

**ML‑инженер: «от данных к обученной модели»**

Вход в систему → «Датасеты и сцены» → Создать новый датасет → выбрать сцены → сохранить → Перейти в «Модели и обучение» → создать эксперимент обучения → Указать модель, датасет, гиперпараметры → запустить → Следить за метриками на странице эксперимента → по завершении модель помечается как «готова к инференсу».

**Аналитик ДЗЗ: «от сцены к картам сегментации»**

Вход → «Инференс» → «Запустить инференс» → Выбрать модель и сцену/датасет → запустить → Открыть карточку запуска → просмотреть исходное изображение и маску в наложении, переключить палитру классов, посмотреть IoU/F1, экспортировать GeoTIFF/PNG.

### Администратор: "управление доступом"

Вход → «Администрирование» → «Пользователи» → Создать пользователя, назначить роль (ML‑инженер/аналитик/админ) → сохранить.

Статусные чипы (completed / running / failed) с цветовой маркировкой реализуют принцип «Помощь в распознавании ошибок».

###6.2. Применение законов проектирования

Закон Фиттса — «Время достижения цели зависит от расстояния и размера элемента»

Основные действия каждого экрана помещены в предсказуемые, крупные зоны, доступные без точного прицеливания:

    Кнопка «Создать датасет» (экран 2) и «Запустить инференс» (экран 3) расположены в верхней части области контента — зона, куда пользователь смотрит в первую очередь.

    Кнопки действий имеют минимальную высоту 32 px (padding + font-size) и горизонтальные поля, чтобы зона клика была шире видимого текста.

    Левый сайдбар с пунктами навигации фиксирован и не требует прокрутки — все разделы достижимы с любой точки экрана одним движением мыши влево.

    Кнопки «Скачать маску» и «Скачать отчёт» на экране «Инференс» сгруппированы рядом в нижней части боковой панели — там, куда взгляд аналитика падает после анализа метрик.

Закон Хика — «Время принятия решения растёт с числом альтернатив»

Количество конкурирующих вариантов на каждом экране намеренно ограничено:

    На каждом экране одна первичная кнопка (акцентный цвет #00897B) и не более двух вторичных — пользователь мгновенно понимает, что делать дальше.

    Навигация скрыта в сайдбаре и не дублируется внутри контента — меню не конкурирует с рабочей областью.

    Вкладки «Датасеты» / «Сцены» на экране 2 делят один длинный список на два коротких, снижая число элементов для обработки вдвое.

    Фильтры в левой панели сгруппированы по типу (тип датасета, наличие разметки, источник) — три группы по 2–3 чекбокса, а не плоский список из 8+ элементов.

Закон Миллера — «Человек удерживает в кратковременной памяти 7 ± 2 элемента»

Информация разбита на блоки, каждый из которых не превышает 7 элементов:

    KPI на «Обзоре» — 4 карточки (меньше нижней границы 5, что позволяет охватить все сразу).

    Таблицы на «Обзоре» показывают последние 3 строки (не все записи) — этого достаточно для быстрой оценки состояния.

    Боковая панель на экране «Инференс» разделена на 4 смысловых блока: «Информация о запуске», «Качество сегментации», «Площадь по классам», «Действия» — каждый блок содержит 2–4 строки.

    Список классов сегментации ограничен 4–6 записями; при большем количестве классов отображаются топ-5 по площади с кнопкой «Показать все».

###6.3. Улучшения относительно аналогов

Улучшение 1. Отсутствие единой точки входа в состояние системы

Проблема в аналогах: В типичных ML-платформах (Google Vertex AI, AWS SageMaker) для понимания текущего состояния проекта нужно последовательно обходить разные разделы (Jobs → Datasets → Models), каждый из которых имеет отдельный дэшборд.

Наше решение: Экран «Обзор» агрегирует 4 KPI-карточки и два последних списка (эксперименты + инференсы) в одном месте. ML-инженер и аналитик видят всё нужное без навигации по разделам — одним взглядом.

Улучшение 2. Смешивание данных, моделей и инференса в одном интерфейсе

Проблема в аналогах: В QGIS и ArcGIS Pro инструменты работы с данными, обучения моделей и применения сегментации перемешаны в панелях плагинов. Пользователь долго ищет нужный инструмент среди десятков панелей.

Наше решение: Навигация разбита на четыре чётких раздела, точно повторяющих этапы ML-пайплайна: «Датасеты и сцены» → «Модели и обучение» → «Инференс» → «Администрирование». Переключение между этапами занимает один клик по сайдбару.

Улучшение 3. Непонятный результат инференса — маска без контекста

Проблема в аналогах: В ряде инструментов (например, специализированные Jupyter-ноутбуки с IPyWidgets) результат сегментации отображается как статичное изображение маски без наложения на исходную сцену и без сопроводительных метрик.

Наше решение: Просмотрщик на экране «Инференс» поддерживает три режима: «Исходное изображение», «Маска», «Наложение» с ползунком прозрачности. Рядом отображаются Mean IoU, Mean F1 и процентное распределение классов — аналитик сразу понимает качество и содержание результата.

Улучшение 4. Отсутствие управления ролями и доступом

Проблема в аналогах: Большинство инструментов для работы с ДЗЗ (SNAP, OTB, скрипты на Python) не предусматривают многопользовательский режим — у всех пользователей один и тот же уровень доступа, что создаёт риски при командной работе.

Наше решение: Экран «Администрирование» предоставляет управление тремя ролями (ML Engineer, RS Analyst, Admin) с чётко описанными правами. Аналитик не может случайно запустить обучение модели, а инженер не имеет доступа к управлению аккаунтами.

Улучшение 5. Неочевидная фильтрация и поиск датасетов

Проблема в аналогах: В MLflow и DVC интерфейс показывает плоский список всех артефактов без визуальной группировки по типу (размеченный / неразмеченный) или статусу разметки. Найти нужный датасет затруднительно при большом числе записей.

Наше решение: На экране «Датасеты и сцены» реализована постоянно видимая левая панель фильтров с группами «Тип датасета», «Наличие разметки» и «Источник». Фильтры применяются мгновенно без перезагрузки, а текущее число найденных записей отображается в строке над таблицей — пользователь видит результат фильтрации сразу.

7.1. Сетка и отступы

Базовая единица сетки: 4 px

Все отступы, размеры и расстояния кратны 4 px. Это упрощает контроль консистентности при реализации и позволяет использовать единую систему токенов.
| Токен    | Значение | Применение                                            |
| -------- | -------- | ----------------------------------------------------- |
| space-1  | 4 px     | Зазор между иконкой и текстом, внутренний отступ чипа |
| space-2  | 8 px     | Отступ внутри компактных элементов (badge, tag)       |
| space-3  | 12 px    | Padding строки таблицы по вертикали                   |
| space-4  | 16 px    | Padding карточки, отступы внутри формы                |
| space-5  | 20 px    | Внутренний padding основного контента                 |
| space-6  | 24 px    | Расстояние между секциями внутри страницы             |
| space-8  | 32 px    | Расстояние между крупными блоками                     |
| space-12 | 48 px    | Верхний/нижний padding разделов                       |

Колонки:

    Макет страницы: 12-колоночная CSS Grid с шириной колонки ~80 px и зазором 16 px при максимальной ширине контента 1280 px.

    Сайдбар занимает 220 px (фиксированно, вне колоночной системы), оставшиеся колонки используются контентом.

    KPI-карточки: 4 колонки (по 3 из 12 на каждую), gap 10 px.

    Блок с двумя панелями («Последние эксперименты» + «Последние инференсы»): пропорция 1.1 / 1, gap 14 px.

Отступы:

    По краям страницы: 16 px (mobile), 20 px (tablet), 24 px (desktop).

    Внутри области контента: padding 20 px сверху/снизу, 20 px слева/справа.

    Между сайдбаром и контентом: border-right 1 px + логический gap.

7.3. Анимации и переходы

Базовые принципы:

    Все анимации функциональны — обозначают изменение состояния или привлекают внимание к важному событию.

    Никаких декоративных анимаций без смысловой нагрузки.

    Соблюдается prefers-reduced-motion: при включённом системном флаге все анимации заменяются мгновенными переходами.
