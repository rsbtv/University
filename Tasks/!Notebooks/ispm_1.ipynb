# **Задание №1. Определение концептуальных основ дипломного проекта в области информационных систем**



В рамках дисциплины "Управление проектами информационных систем" представьте развернутое концептуальное обоснование вашего проекта, ответив на следующие вопросы:


#### **1. Обоснование необходимости разработки**


Для кого предназначено разрабатываемое Вами решение и какие потребности целевой аудитории оно призвано удовлетворить?




*Обоснуйте актуальность решаемой проблемы для конкретных заинтересованных сторон.*

---

Проект нацелен на автоматическую сегментацию спутниковых снимков для построения карт типов поверхности (лес, вода, поля, застройка и др.) на уровне крупных территорий, включая задачи градостроительства и строительного мониторинга. Заинтересованными сторонами являются органы власти, отвечающие за территориальное планирование и контроль застройки, девелоперские и строительные компании, а также организации агросектора, которым регулярно требуются сопоставимые карты землепользования. В качестве базовой модели выбрана HyenaPixel: ее разработчиками она описана как архитектура, использующая свёртки с очень большим охватом и учитывающая контекст всего изображения, но её применение к данным ДЗЗ в доступной литературе не встречается, что создает новизну выполняемой задаче.

#### **2. Ожидаемые результаты проекта**



Опишите конкретные результаты, которые планируется получить по завершении проекта. Какой программный продукт, методика или информационная система будет создан/создана в результате вашей работы?

---

Ожидается получение методики сегментации данных ДЗЗ на основе HyenaPixel, в которой будут заданы конкретные настройки модели под спутниковые снимки: размеры тайлов, используемые спектральные каналы, глубина и конфигурация слоёв. В методику также войдут шаг предобучения на неразмеченных спутниковых снимках, шаг дообучения на размеченных масках и описанная процедура проверки качества на выбранных наборов данных, например на датасетах с классами землепокрытия. Технически результат будет представлен в виде репозитория с исходным кодом модели, конфигурационными файлами запуска, сохранёнными файлами с обученными параметрами и протокола экспериментов, в котором фиксируются использованные датасеты, настройки и полученные числовые показатели.

#### **3. Методология использования**


Каким образом предполагается практическое применение разрабатываемого решения? Опишите основные сценарии взаимодействия пользователей с разрабатываемым решением.

Использование решения предполагает типовой цикл: сначала спутниковые снимки Sentinel‑2 и при необходимости Sentinel‑1 загружаются из открытых хранилищ, затем к ним применяется предобработка с маскированием облаков, выравниванием по единой сетке, приведением разрешений и нарезкой на тайлы в формате GeoTIFF. После этого подготовленные тайлы подаются в модель, которая выдаёт для каждого снимка карту классов, и эти карты сохраняются как отдельные растровые файлы. Полученные растры подключаются в ГИС и могут использоваться для анализа землепокрытия и изменений, органами власти для обновления карт и контроля застройки, строительными и девелоперскими компаниями, для отслеживания динамики освоения площадок и прилегающих территорий.

#### **4. Входные данные и их источники**


Какие типы данных будут использоваться в качестве входных данных для вашего решения? В каком формате они будут представлены и из каких источников планируется их получение?

Входными данными выступают мультиспектральные снимки Sentinel‑2 уровня Level‑2A с 12 спектральными каналами и пространственным разрешением порядка 10–20 метров. Эти данные приводятся к виду геопривязанных тайлов размером примерно 256×256–512×512 пикселей в формате GeoTIFF, пригодных для пакетной обработки моделью. Источником неразмеченных данных для предобучения служат открытые хабы Copernicus, USGS и аналогичные сервисы, тогда как для обучения и проверки сегментации используются размеченные наборы, такие как SEN12MS и FLAIR, где уже заданы классы землепокрытия и соответствующие маски.

#### **5. Выходные данные и практическая значимость**


Какие результаты будет формировать Ваше решение на выходе?

На выходе формируются растровые файлы в формате GeoTIFF, в которых каждому пикселю соответствует определённый класс поверхности (лес, вода, поля, застройка и другие категории), и при необходимости из этих растров дополнительно могут быть построены векторные файлы с контурами объектов в форматах вроде shapefile или GeoPackage. Дополнительно планируется создание таблицы с числовыми показателями точности и времени обработки для разных участков и настроек, которые позволяют сравнивать конфигурации модели и режимы её использования.

---

Опишите конкретно и без "воды" пользу (практическую ценность), которую заинтересованные стороны получат от использования этих результатов.

Эти данные могут напрямую использоваться органами власти для актуализации карт землепользования и контроля строительной деятельности, строительными и девелоперскими компаниями — для оценки фактического хода застройки и освоения территорий, аграрными структурами — для расчёта площадей сельхозкультур, а исследовательскими центрами — для постановки и проверки задач в области анализа пространственных данных.
Возможные примеры:

2) строительные и девелоперские компании — контроль развития строительных площадок и прилегающей инфраструктуры по спутниковым данным;

3) агросектор — расчёт площадей сельхозкультур и мониторинг изменений посевов;
